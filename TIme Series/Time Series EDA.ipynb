# Time Resampling
The process of changing the frequency of the time series data. Time series data often come with timestamps that may not be at a desired frequency (e.g., irregular intervals). Resampling allows you to organize or aggregate the data into a different frequency, making it more suitable for analysis or visualization.


```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A')
# The rule is 'A', which stands for annually (year-end frequency).
```

```python
# there are diff diff property of resample
# there ar many rule A ,B , C
df_tesla.resample(rule='A').min()
# By min we can find out the minimun of every column on that purticular year
```

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
# by using open we can plot the maximum value from the date
```

```python
# For quaterly end frequency (january april and so on)
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
# from above link you can found all rule
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
df_tesla.resample(rule='QS').max()['Open'].plot()
```

The 'BA' frequency stands for Business Year End Frequency. It is specifically designed to represent the end of a business year, typically ending on the last weekday of the month (business month end). This is useful when working with financial or business-related time series data.

```python
## Buisness End Frequency
df_tesla.resample(rule='BA').max()['Open'].plot()
```

```python
df_tesla.resample(rule='BA').max()
# last day of every year
```

```python
# Buisness quater
df_tesla.resample(rule='BQS').max()
```

```python
## Plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# M stands for monthly data
df_tesla['Open'].resample(rule='M').mean()
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

# Rolling

### The rolling method is used to smoothen the curve. A rolling window calculation involves applying a function (such as mean, sum, std, etc.) to a specified window of consecutive data points in a time series.
### Can also be called Mean Average 

So, rolling(10) specifies a rolling window of size 10, meaning that for each data point, the calculation considers the average of the current value and the previous 9 values (totaling 10 values in the rolling window). This type of calculation is often used for smoothing out short-term fluctuations and highlighting longer-term trends in time series data, in this case, the 'High' prices of Tesla stock.

```python
## Some Aggregate functon such as Rolling
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['Open: 30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

 the code adds a new column to the DataFrame ('Open: 30 days rolling') that represents the 30-day rolling mean of the opening prices of Tesla stock. This calculated column is useful for smoothing out short-term fluctuations and revealing longer-term trends in the opening prices. It provides a moving average over a 30-day window, helping to identify trends in the data

```python
 df_tesla.head(31)
```

```python
df_tesla[['Open','Open: 30 days rolling']].plot(figsize=(12,5))
```
